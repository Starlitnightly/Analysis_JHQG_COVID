```python
def get_inte_target(inte_name):
    data1={
      "v": inte_name,
      "label": "Ingredient",
      "key_id": inte_name,
      "func_name": "detail_api"
    }
    #在get请求内，添加user-agent
    session = requests.session()
    ret = session.post(url = "http://herb.ac.cn/chedi/api/?", headers = headers, data=json.dumps(data1),cookies=cookies)
    if(ret.status_code!=200):
        return 'error'
    
    
    a=json.loads(ret.content)
    res=[]
    for i in range(len(a['drug_paper_target'])):
        if i==0:
            continue
        res.append(a['drug_paper_target'][i][2])
    '''
    for i in range(len(a['ingredient_target'])):
        if i==0:
            continue
        res.append(a['ingredient_target'][i][1])
    '''
    res=list(set(res))
    return res
```

```python
def plot_network(herb,out_path):
    G = nx.Graph()
    herb_jvhua=herb
    for i in herb_jvhua.index:
        G.add_edge(herb_jvhua.loc[i,'herb_id'],herb_jvhua.loc[i,'target'])
        
    nodes=[]
    G_degree=dict(G.degree(G.nodes()))
    num_aging=0
    num_protein=0
    for i in G.nodes:
        no={}
        no['name']=i
        no['symbolSize']=G_degree[i]/5
        if i in herb_jvhua['herb_id'].values:
            no['category']=herb_jvhua.loc[herb_jvhua['herb_id']==i,'herb_type'].values[0]
        #no['is_fixed']=True
        else:
            no['category']='Protein'
        num_protein+=1
                
            #p[i]=1
        nodes.append(no)
    #print(out_path.split('.')[0],'\nProp-Aging:{0}%,({1},{2})\n'.format((num_aging/(num_aging+num_protein))*100,num_aging,num_protein+num_aging))

    links = []
    for i in herb_jvhua.index:
        links.append({"source": herb_jvhua.loc[i,'herb_id'], "target": herb_jvhua.loc[i,'target']})
    
    categories=[{
                    "name": "Zhebeimu",
                    "itemStyle": {
                        "normal": {
                            "color": '#c2c2c2', #公司颜色为蓝
                            "borderColor":"#5182ab",
                            "borderWidth":1.8
                        }
                    }
                },{
                    "name": "Qinghao",
                    "itemStyle": {
                        "normal": {
                            "color": '#a51616', #供应商颜色为红
                            "borderColor":"#5182ab",
                            "borderWidth":1.8
                        }
                    }
                }, 
                  {
                    "name": "terpenes",
                    "itemStyle": {
                        "normal": {
                            "color": '#357517', #供应商颜色为红
                            "borderColor":"#5182ab",
                            "borderWidth":1.8
                        }
                    }
                },{
                    "name": "flavonoids",
                    "itemStyle": {
                        "normal": {
                            "color": '#e54746', #供应商颜色为红
                            "borderColor":"#5182ab",
                            "borderWidth":1.8
                        }
                    }
                },{
                    "name": "steroids",
                    "itemStyle": {
                        "normal": {
                            "color": '#f6aa00', #供应商颜色为红
                            "borderColor":"#5182ab",
                            "borderWidth":1.8
                        }
                    }
                },{
                    "name": "fatty acids",
                    "itemStyle": {
                        "normal": {
                            "color": '#4f61c7', #供应商颜色为红
                            "borderColor":"#5182ab",
                            "borderWidth":1.8
                        }
                    }
                },{
                    "name": "Mixed",
                    "itemStyle": {
                        "normal": {
                            "color": '#c2c2c2', #供应商颜色为红
                            "borderColor":"#5182ab",
                            "borderWidth":1.8
                        }
                    }
                }
        ]
    categories=[{"name": c,
                    "itemStyle": {
                        "normal": {
                            "color": color,
                            "borderColor":"#5182ab",
                            "borderWidth":1.8
                        }
                    }} for c,color in zip(list(set(herb['herb_type']))+['Protein'],ov.utils.sc_color)]

    c = (
        Graph(init_opts=opts.InitOpts(#bg_color='rgba(255,250,205,0.2)',
                                    width='1600px',
                                    height='800px',
                                    page_title='page',
                                    ))

        .add("", nodes, links,
             categories = categories,
             is_draggable=True,
             gravity=0.1,
             #is_roam= True,
             #layout='circular',
              #graph_layout="force",
              repulsion=150,
              #graph_edge_symbol=['cricle'],

            )
        .set_global_opts(title_opts=opts.TitleOpts(title=out_path))
        .render(out_path)
    )
    return G
```
